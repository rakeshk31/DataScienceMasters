```python
# # What is Flask Framework? What are the advantages of Flask Framework?
# Flask is a micro web framework written in Python. It is designed to be lightweight and easy to use, making it a popular choice for building web applications and APIs. Here are some of the advantages of using the Flask framework:

# Simplicity: Flask follows a minimalist design philosophy, providing only the essentials for web development. It has a small core and modular architecture, allowing developers to choose and add extensions as needed. This simplicity makes it easy to understand, learn, and get started with Flask.

# Flexibility: Flask does not enforce any specific project structure or dependencies, giving developers the freedom to structure their applications as they see fit. It allows you to choose the components you need, such as database integration, form handling, or authentication, by leveraging various Flask extensions. This flexibility makes Flask adaptable to a wide range of use cases.

# Lightweight: Flask is lightweight and has a small footprint, which means it has fewer dependencies and minimal overhead. This makes Flask efficient in terms of resource usage, making it suitable for building applications with limited resources or those that need to handle a large number of concurrent requests.

# Extensibility: Flask provides a rich ecosystem of extensions that can be easily integrated into your application. These extensions cover a wide range of functionalities, including database integration, authentication, form handling, caching, and more. The modular design of Flask allows you to pick and choose the extensions you need, enabling you to extend the capabilities of your application without unnecessary bloat.

# Testing and Debugging: Flask provides a built-in development server that supports interactive debugging. It also has a testing framework that makes it easy to write and run unit tests for your application. These features help streamline the development process, making it easier to identify and fix issues during development.

# Python Integration: Being a Python framework, Flask leverages the power and versatility of Python. It allows developers to use the vast range of Python libraries and modules, making it easier to integrate existing Python code into Flask applications.
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?

# In Flask, app routing refers to the process of mapping URLs to specific functions or views in your web application. It allows you to define different routes (URL patterns) that correspond to specific actions or pages in your application. When a user requests a specific URL, Flask's routing mechanism matches the URL to the corresponding route and triggers the associated function, which generates the response to be sent back to the user's browser.

# App routes in Flask are defined using the @app.route() decorator, which is a way to associate a URL with a function. 
```
